```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
common_size = (600, 800)
```

```python
image1 = cv2.imread('Car2.jpeg')
image1 = cv2.resize(image1, common_size)
plt.figure()
plt.imshow(image1)
plt.show()
```

```python
image2 = cv2.imread('Bike2.jpg')
image2 = cv2.resize(image2, common_size)
plt.figure()
plt.imshow(image2)
plt.show()
```

## Image Fusion

```python
alpha = 0.6
beta = 0.4
fusion_result = cv2.addWeighted(image1, alpha, image2, beta, 0)
```

```python
plt.figure()
plt.imshow(fusion_result)
plt.show()
```

## Overlapping of images

```python
overlap_result = image1.copy()
height, width, _ = image2.shape
overlap_region = overlap_result[0:height, 0:width]
blended_overlap = cv2.addWeighted(overlap_region, 0.5, image2, 0.5, 0)
overlap_result[0:height, 0:width] = blended_overlap
```

```python
plt.figure()
plt.imshow(overlap_result)
plt.show()
```

## Separate the two images.

```python
image1_separated = np.divide(fusion_result, alpha + beta) * alpha
image2_separated = np.divide(fusion_result, alpha + beta) * beta
```

```python
plt.figure()
plt.imshow(image1_separated.astype(np.uint8))
plt.show()
```

```python
plt.figure()
plt.imshow(image2_separated.astype(np.uint8))
plt.show()
```
